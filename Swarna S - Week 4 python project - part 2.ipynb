8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol',
'total prod', 'stocks', 'price per lb', 'prodvalue'.

```python
column=['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
correlation_matrix =data[column].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```

Observation:
   The heatmap distribution is a highest positive correlation value is a 'numcol' to 'totalprod' which is a expected to highest number of colonies produce more honey.strongly positive correlation in a 'total prod' to 'prod value'.the negative correlation is a 'yield percol','stocks''prodvalue'.which makes since increase in production surplus to honey.and also decrease in price.
